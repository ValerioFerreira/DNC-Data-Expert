```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('/content/heart.csv')
```

```python
df.head()
```

```python
df.corr()
```

```python
X = df.drop(columns = ['target'])
y = df['target']
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
pca_fit = pca.fit_transform(X, y)

df_final = pd.DataFrame(data = pca_fit, columns = ['P1', 'P2'])
df_final['hue'] = df['target']
sns.scatterplot(data=df_final,x='P1',y='P2',hue='hue')
```
